```python
import random
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
# Step 1: Load pre-trained language model and tokenizer
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
# Step 2: Split document into chunks
def split_into_chunks(document, chunk_size):
    sentences = document.split(". ")  # You may need a more sophisticated sentence tokenizer
    chunks = [sentences[i:i + chunk_size] for i in range(0, len(sentences), chunk_size)]
    return chunks
```

```python
# Step 3: Randomly select a percentage of chunks
def select_random_chunks(chunks, percentage):
    num_chunks_to_select = int(len(chunks) * (percentage / 100))
    selected_chunks = random.sample(chunks, num_chunks_to_select)
    return selected_chunks
```

```python
# Step 4: Generate summary from selected chunks
def generate_summary(selected_chunks):
    summary = ""
    for chunk in selected_chunks:
        input_text = " ".join(chunk)
        input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
        output = model.generate(input_ids, max_length=150, num_return_sequences=1, no_repeat_ngram_size=2, top_p=0.92)
        generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
        summary += generated_text + " "
    return summary
```

```python
# Step 5: Prompt language model to find incomplete parts
def complete_summary(summary):
  # Use GPT-2 model to generate additional content and complete the summary
  input_ids = tokenizer.encode(summary, return_tensors="pt", max_length=512, truncation=True)
  # Check if the input tensor is not empty
  if input_ids.numel() > 0:
    output = model.generate(input_ids, max_length=150, num_return_sequences=1, no_repeat_ngram_size=2, top_p=0.92)
    # Check if the output tensor is not empty
    if output.numel() > 0:
      generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
      return generated_text

```

```python
# Step 6: Provide relevant and nearby chunks to improve the summary
def provide_relevant_chunks(original_chunks, selected_chunks):
    # Identify nearby chunks based on the original document structure
    # You may need to define a metric to determine the relevance of a chunk

    # For simplicity, let's just select the adjacent chunks for each selected chunk
    improved_chunks = []
    for chunk in selected_chunks:
        chunk_index = original_chunks.index(chunk)
        if chunk_index > 0:
            improved_chunks.append(original_chunks[chunk_index - 1])
        improved_chunks.append(chunk)
        if chunk_index < len(original_chunks) - 1:
            improved_chunks.append(original_chunks[chunk_index + 1])

    return improved_chunks
```

```python
def read_document_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        document = file.read()
    return document
```

```python
document = read_document_from_file("/content/drive/MyDrive/ColabNotebooks/test.txt")
chunk_size = 1
percentage_to_select = 50

chunks = split_into_chunks(document, chunk_size)
selected_chunks = select_random_chunks(chunks, percentage_to_select)
summary = generate_summary(selected_chunks)
completed_summary = complete_summary(summary)
improved_chunks = provide_relevant_chunks(chunks, selected_chunks)
```
